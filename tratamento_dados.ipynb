# Importing libraries

```python
# import 'pandas' library
import pandas as pd
```

# Reading file

```python
# reading 'xls' file
df_vendas = pd.read_csv('vendas.xls')

# show dataframe
df_vendas
```

# Verifications and validations

```python
# verify the data types
df_vendas.dtypes
```

```python
# convert 'data' to 'datatype'
df_vendas['data'] = pd.to_datetime(df_vendas['data'])
```

```python
# verify new type from 'data'
df_vendas.dtypes
```

```python
# check for missing values
if df_vendas.isna().any().any():
  df_vendas = df_vendas.fillna(0)
```

# Columns manipulations

```python
# rename the 'data' column
df_vendas = df_vendas.rename(columns={'data_venda': 'data'})
df_vendas
```

```python
# create a new column
df_vendas['receita_total'] = df_vendas['quantidade'] * df_vendas['preco_unitario']
df_vendas
```

```python
# Create 'mês' column
df_vendas['mês'] = df_vendas['data'].dt.month
df_vendas
```

# Data filtering

```python
# Data filtering by 'regiao'
df_sudeste = df_vendas[df_vendas['regiao'] == 'Sudeste']
df_nordeste = df_vendas[df_vendas['regiao'] == 'Nordeste']
df_norte = df_vendas[df_vendas['regiao'] == 'Norte']
df_sul = df_vendas[df_vendas['regiao'] == 'Sul']
```

```python
# Show filtering data by 'regiao'
df_nordeste
```

```python
# Calculates average sales per product
df_average_sales_per_product = df_vendas.groupby('produto')['quantidade'].mean()
df_average_sales_per_product
```

# Pivot table

```python
# total quantity sold by product and seller
df_pivot_product_by_seller = df_vendas.pivot_table(
  index='produto',
  columns='vendedor',
  values='quantidade',
  aggfunc='sum'
)
df_pivot_product_by_seller
```

```python
# total products by region
df_pivot_product_by_region = df_vendas.pivot_table(
  index='produto',
  columns='regiao',
  values='quantidade',
  aggfunc='sum'
)
df_pivot_product_by_region
```

```python
# sales by month
df_sales_by_month = df_vendas.pivot_table(
  index=['produto', 'regiao'],
  columns='mês',
  values='quantidade',
  aggfunc='sum'
)

df_sales_by_month
```

# Functions

```python
# Function to categorize sellers based on quantity sold
def categorize_seller(qntd):
  if qntd > 100:
    return 'Vendedor Destaque'
  elif qntd > 50:
    return 'Vendedor Médio'
  else:
    return 'Vendedor Iniciante'
  
# Calling function and create a new column
df_vendas['categoria_vendedor'] = df_vendas['quantidade'].apply(categorize_seller)

# Show the updated dataframe
df_vendas
```
